```python
import shutil
import os
import cv2
import numpy as np
```

```python

def move_file(folder_path):
    file_list=os.listdir(folder_path)

    file_names=[]
    for f in file_list:
        if f[:17] not in file_names:
            file_names.append(f[:17])

    for file in file_names:
        file_path=os.path.join(folder_path,file)
        if os.path.isdir(file_path)==0:
            os.mkdir(file_path)
        
        raw_predict_result=os.path.join(file_path,'raw_predict')
        blend_result=os.path.join(file_path,'blend')
        if os.path.isdir(raw_predict_result)==0:
            os.mkdir(raw_predict_result)
        if os.path.isdir(blend_result)==0:
            os.mkdir(blend_result)

    image_list=list(set(file_list)-set(file_names))

    for folder in file_names:
        for file in image_list:
            source_path=os.path.join(folder_path,file)
        
            if (file.find(folder)!=-1) & (file.find('blend')==-1):
                destination_path=os.path.join(os.path.join(os.path.join(folder_path,folder),'raw_predict'),file)
                shutil.move(source_path, destination_path)
                print(f'move {source_path} to {destination_path}')

            elif (file.find(folder)!=-1) & (file.find('blend')!=-1):
                destination_path=os.path.join(os.path.join(os.path.join(folder_path,folder),'blend'),file)
                shutil.move(source_path, destination_path)
                print(f'move {source_path} to {destination_path}')
            
            else:
                continue
```

```python
folder_path="E:/불량검출-태광-결과/250117_seat_indenter"
move_file(folder_path)
```

```python
def merge_patches_to_image(slice_folder,patch_size,num_patches_x,num_patches_y,save_path):
    """
    slice_folder = ['./img1.png','./img1.png','./img1.png','./img1.png',...]
    patch_size = (512, 512, 3)  # 개별 패치의 크기
    num_patches_x = 8  # 가로 패치 개수
    num_patches_y = 6  # 세로 패치 개수
    """
    full_image_size = (num_patches_y * patch_size[0], num_patches_x * patch_size[1], patch_size[2])

    # 빈 이미지 생성
    full_image = np.zeros(full_image_size, dtype=np.uint8)

    patches=[cv2.imread(i) for i in slice_folder]
  
    # 패치 합치기
    count = 0
    for y in range(num_patches_y):
        for x in range(num_patches_x):
            start_y = y * patch_size[0]
            start_x = x * patch_size[1]
            full_image[start_y:start_y + patch_size[0], start_x:start_x + patch_size[1], :] = patches[count]
            count += 1
            
    cv2.imwrite(save_path, full_image)
    print(f'save merge image in {save_path}')
    return 0

```

```python
patch_size = (512, 512, 3)
num_patches_x = 8
num_patches_y = 6

folder_list=os.listdir("E:/불량검출-태광-결과/250117_seat_indenter")
print(folder_list)
for i in folder_list:
    path=f"H:/AI-Data/New_Rule/taegwang/fitting_valve/241023 BMT_gRAND vISION_naming/outside_anomaly/result/train_patch_img_512_epoch_19_bad_outline/{i}"
    save_path=os.path.join(path,f'{i}_merge.png')
    slice_folder=sorted(os.listdir(os.path.join(path+'/blend')))
    slice_folder=list(map(lambda x: os.path.join(os.path.join(path,'blend'),x),slice_folder))
    merge_patches_to_image(slice_folder,patch_size,num_patches_x,num_patches_y,save_path)
```
