```python
import os
import glob
import cv2
import numpy as np
from PIL import Image
```

```python

root_dir='E:/Git/Hyper/sheat_label_two'

images = sorted(glob.glob(root_dir+"/*"))
images
```

```python
for i,_ in enumerate(images):
    mask_path=images[i]
    mask = cv2.imread(mask_path, cv2.IMREAD_GRAYSCALE)
    mask=np.where(mask>1,255,0).astype(np.uint8) 
    
    image = Image.fromarray(mask)
    # image.show()
    image.save(mask_path)
```

```python

image.show()

# 이미지 파일로 저장
#
```
