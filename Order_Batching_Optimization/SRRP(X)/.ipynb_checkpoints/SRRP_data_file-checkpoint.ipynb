```python
import pandas as pd
import numpy as np
import os
import warnings
warnings.filterwarnings('ignore')
pd.options.display.max_rows = 9999
```

```python
def createFolder(directory):
    try:
        if not os.path.exists(directory):
            os.makedirs(directory)
    except OSError:
        print ('Error: Creating directory. ' +  directory)


def transform_df(df):

    df.columns=['name']

    inventory='Initial item inventory level'
    max_capacity='Maximum inventory capacity per item'
    replenishment='Reserve storage replenishment rate per item and per period'
    demand='Forward storage demand rate per item and per period'
    distance='Distance matrix'

    inventory_index=df.query("name=='Initial item inventory level'").index[0]
    max_capacity_index=df.query("name=='Maximum inventory capacity per item'").index[0]
    replenishment_index=df.query("name=='Reserve storage replenishment rate per item and per period'").index[0]
    demand_index=df.query("name=='Forward storage demand rate per item and per period'").index[0]
    distance_index=df.query("name=='Distance matrix'").index[0]

    #file_name
    title=df[:4]

    #inventory file
    inventory_df=df[inventory_index+1:max_capacity_index]
    inventory_df.columns=[inventory]
    inventory_df[inventory]=inventory_df[inventory].astype(dtype='int')
    inventory_df.reset_index(drop=True, inplace=True)

    #capacity file
    max_capacity_df=df[max_capacity_index+1:replenishment_index]
    max_capacity_df.columns=[max_capacity]
    max_capacity_df[max_capacity]=max_capacity_df[max_capacity].astype(dtype='int')
    max_capacity_df.reset_index(drop=True, inplace=True)

    #replenishment file
    replenishment_df=df[replenishment_index+1:demand_index].apply(lambda x:x.str.split('\t'))
    replenishment_df.columns=[replenishment]
    replenishment_df.reset_index(drop=True, inplace=True)
    for i in list(replenishment_df.index):
        rep=list(list(filter(None, replenishment_df.loc[i][0])))
        replenishment_df.loc[i,replenishment]=list(map(float,rep))

    #demand file
    demand_df=df[demand_index+1:distance_index].apply(lambda x:x.str.split('\t'))
    demand_df.columns=[demand]
    demand_df.reset_index(drop=True, inplace=True)
    for i in list(demand_df.index):
        dem=list(list(filter(None, demand_df.loc[i][0])))
        demand_df.loc[i,demand]=list(map(float,dem))

    #distnace file
    distance_df=df[distance_index+1:].apply(lambda x:x.str.split('\t'))
    distance_df.columns=[distance]
    distance_df.reset_index(drop=True,inplace=True)
    for i in list(distance_df.index):
        dis=list(list(filter(None, distance_df.loc[i][0])))
        distance_df.loc[i,distance]=list(map(float,dis))

    dist=[]
    for i in distance_df[distance]:
        dist.append(i)
    distance_nd=np.array(dist)
    distance_nd.shape
    
    return title,inventory_df,max_capacity_df,replenishment_df,demand_df,distance_nd
```

```python
path='./SRRP-instances/Small instances/'
file_list=os.listdir(path)

for f in file_list:
    createFolder(path+f[:-4])
    df=pd.read_csv(path+f,header=None,engine='python')
    title,inventory_df,max_capacity_df,replenishment_df,demand_df,distance_nd=transform_df(df)
    title.to_csv(path+f[:-4]+'/meta_info.csv',index=False)
    inventory_df.to_csv(path+f[:-4]+'/inventory.csv',index=False)
    max_capacity_df.to_csv(path+f[:-4]+'/max_capacity.csv',index=False)
    replenishment_df.to_csv(path+f[:-4]+'/replenishment.csv',index=False)
    demand_df.to_csv(path+f[:-4]+'/demand.csv',index=False)
    np.save(path+f[:-4]+'/distance.npy',distance_nd)
```
